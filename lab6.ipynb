```python
#task1 
import heapq

def heuristic(a,b):
    return abs(a[0]-b[0])+ abs(a[1]-b[1])

def get_neighbors(pos,maze):
    x, y = pos
    neighbors = []

    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    for dx, dy in directions:
        nx,ny = x+dx, y+dy

        if 0 <= nx < len(maze) and 0 <= ny < len(maze[0]):
            if maze [nx][ny] !='#':
                neighbors.append((nx,ny))
    return neighbors

def best_fs(maze,start,goal):
    pq = []
    heapq.heappush(pq,(heuristic(start,goal),start))
    visited=set()
    parent = {}

    while pq:
        _, current = heapq.heappop(pq)

        if current == goal:
            path = []
            while current in parent:
                path.append(current)
                current = parent[current]
            path.append(start)
            path.reverse()
            return path

        visited.add(current)

        for neighbor in get_neighbors(current,maze):
            if neighbor not in visited:
                parent[neighbor] = current
                heapq.heappush(pq,(heuristic(neighbor,goal),neighbor))
    return None

def multi_goal_navigation(maze,start,goals):

    current = start
    full_path = []

    goals = set(goals)

    while goals:

        goal = min(goals,key=lambda g: heuristic(current,g))

        path = best_fs(maze,current,goal)

        if path is None:
            print("No path to goal")
            return 

        full_path += path[:-1]
        current = goal
        goals.remove(goal)
    full_path.append(current)
    return full_path

maze = [
['S','.','.','#','.'],
['.','#','.','#','.'],
['.','#','.','.','.'],
['.','.','#','G','.'],
['G','.','.','.','.']
]

start = (0,0)
goals = [(3,3),(4,0)]
path = multi_goal_navigation(maze,start,goals)

print("Path covering all goals:")
print(path)
```

```python
#task 3 
import heapq
import math

deliveries = {
"A": {"location":(2,3),"deadline":10},
"B": {"location":(5,2),"deadline":5},
"C": {"location":(6,6),"deadline":15},
"D": {"location":(1,7),"deadline":8}
}

def distance(a,b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

def greedy_delivery(start):
    current_pos=start
    time =0

    visited = set()
    route =[]

    while len(visited) < len(deliveries):

        pq=[]
        for name,data in deliveries.items():
            if name in visited:
                continue
            dist = distance(current_pos,data["location"])
            urgency = data["deadline"] - time
            priority = dist + urgency
            heapq.heappush(pq,(priority,name))
        _, chosen = heapq.heappop(pq)
        visited.add(chosen)
        location = deliveries[chosen]["location"]
        travel = distance(current_pos,location)
        time += travel
        route.append((chosen,time))
        current_pos = location
    return route

route = greedy_delivery((0,0))

print("Delivery Order:")
for r in route:
    print("Deliver to",r[0],"at time",round(r[1],2))
```
