```python
#task1
#DLS
class GoalBasedAgentDLS:
    def __init__(self,graph,goal,depth_limit):
        self.graph=graph
        self.goal=goal
        self.depth_limit=depth_limit
        
    def dls(self,current,depth):
        print(f"Visiting: {current},Depth : {depth}")
        
        if current == self.goal:
            return [current]
        if depth == 0:
            return None
        for neighbour in self.graph.get(current,[]):
            result = self.dls(neighbour,depth-1)
            if result is not None:
                 return [current] + result
        return None
    
    def search(self,start):
        return self.dls(start,self.depth_limit)
    
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}
agent = GoalBasedAgentDLS(graph,goal='E',depth_limit=2)
path = agent.search('A')

print("\n final path",path)
```

```python
#UCS
import heapq

class UtilityBasedAgentUCS:
    def __init__(self,graph,goal):
        self.graph = graph
        self.goal =  goal
    def search(self,start):
        pq=[]
        heapq.heappush(pq,(0,start,[start]))
        
        visited = set()
        
        while pq:
            cost,current,path = heapq.heappop(pq)
            
            if current in visited:
                continue
            print(f"Visiting: {current}, Cost: {cost}")
            
            visited.add(current)
            
            if current == self.goal:
                return path,cost
            
            for neighbour, weight in self.graph.get(current,[]):
                if neighbour not in visited:
                    heapq.heappush(pq,(cost + weight,neighbour,path+[neighbour]))
        return None
    
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}
agent = UtilityBasedAgentUCS(graph,goal='E')
result = agent.search('A')

print("\nFinal Result:",result)
```

```python
#task2
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
n=4
visited = [False] * n
min_cost = float('inf')
best_path = []

def tsp(current,count,cost,path):
    global min_cost,best_path
    
    if count == n and graph[current][0] > 0:
        total_cost = cost + graph[current][0]
        if total_cost < min_cost:
            min_cost = total_cost
            best_path = path + [0]
        return
    
    for i in range(n):
        if not visited[i] and graph[current][i] > 0:
            visited[i]=True
            tsp(i,count + 1,cost + graph[current][i],path + [i])
            visited[i] = False
            
visited[0] = True
tsp(0, 1, 0, [0])
print("Minimum cost : ",min_cost)
print("Best Path ",[city+1 for city in best_path])
```

```python
#task3
#IDDFS on graph
class IDDFS_Graph:
    def __init__(self,graph):
        self.graph = graph
    def dls(self,node,goal,depth,visited):
        if depth<0:
            return False
        print("Visiting :",node)
        
        if node == goal:
            return True
        visited.add(node)
        
        for neighbour in self.graph.get(node,[]):
            if neighbour not in visited:
                if self.dls(neighbour,goal,depth - 1,visited):
                    return True
        return False
    def iddfs(self,start,goal,max_depth):
        for depth in range(max_depth + 1):
            print("\n Depth Limit:",depth)
            visited = set()
            if self.dls(start,goal,depth,visited):
                print("Goal found at depth",depth)
                return True
        return False
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

agent = IDDFS_Graph(graph)
agent.iddfs('A', 'E', 3)
```

```python
#iddfs on tree
class TreeNode:
    def __init__ (self,value):
        self.value = value
        self.children = []
def dls_tree(node,goal,depth):
    if node is None:
        return False
    print("Visiting:",node.value)
    
    if node.value == goal:
        return True
    if depth == 0:
        return False
    for child in node.children:
        if dls_tree(child,goal,depth - 1):
            return True
    return False
def iddfs(root,goal,max_depth):
    for depth in range(max_depth + 1):
        print("\nDepth Limit",depth)
        if dls_tree(root,goal,depth):
            print("Goal found at depth",depth)
            return True
        return False
    
A = TreeNode('A')
B = TreeNode('B')
C = TreeNode('C')
D = TreeNode('D')
E = TreeNode('E')

A.children = [B, C]
B.children = [D, E]

iddfs(A, 'E', 3)
```
