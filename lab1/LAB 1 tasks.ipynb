```python
# Task 3


n = int(input("Enter how many numbers you want to input: "))


numbers = []


for i in range(n):
    value = int(input(f"Enter number {i + 1}: "))
    numbers.append(value)


even_count = 0
odd_count = 0


for num in numbers:
    print(num)
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1


print("Count of even numbers:", even_count)
print("Count of odd numbers:", odd_count)


index = int(input("\nEnter index you want to replace: "))
newVal = int(input("Enter new value: "))

if 0 <= index < n:
    numbers[index] = newVal
    print("Updated list:", numbers)
else:
    print("Invalid index")
```

```python
# task 4
#Create a function that:
# Takes basic salary as parameter
# Calculates:
#o HRA = 10% of salary
#o DA = 5% of salary
# Returns the total salary
# Call the function and display the result

def calc_salary(basic_Sal):
    HRA = 0.1 * basic_Sal
    Da = 0.05 * basic_Sal 
    total_salary = basic_Sal+HRA+Da
    return total_salary
salary = float(input("enter salary : "))
result  = calc_salary(salary)
print("total salary : ",result)
```

```python
#task 5
#Create a function that:
# Accepts a string as input
# Returns a dictionary containing:
#o Total characters
#o Number of uppercase letters
#o Number of lowercase letters
#o Number of digits

def string_analysis(text):
    total_char=len(text)
    uppercase = 0
    lowercase = 0
    digits = 0
    
    for ch in text :
        if ch.isupper():
            uppercase+=1
        elif ch.islower():
              lowercase+=1
        elif ch.isdigit():
              digits +=1
    result = {
        "Total Characters : " : total_char,
        "Upper-Case : "       : uppercase,
        "lower case :"        : lowercase,
        "digits :"             : digits
    }   
    
    return result

string  = input("Enter a string : ")
a = string_analysis(string)
print (a)
```

```python
# Task 6

text = input("Enter a string: ")


longest_substring = ""
current_substring = ""

for i in range(len(text)):
    if i == 0 or text[i] == text[i - 1]:
        current_substring += text[i]
    else:
        current_substring = text[i]

    if len(current_substring) > len(longest_substring):
        longest_substring = current_substring

print("Longest substring:", longest_substring)
print("Its length:", len(longest_substring))
```

```python
#task 7
def find_repeated_numbers(numbers):
    repeated = []

 
    for i in range(len(numbers)):
        count = 0

        
        for j in range(len(numbers)):
            if numbers[i] == numbers[j]:
                count += 1


        if count > 1:
         
            already_added = False
            for k in range(len(repeated)):
                if numbers[i] == repeated[k]:
                    already_added = True
                    break

            if not already_added:
                repeated.append(numbers[i])

    
    return repeated
nums = [1,2,3,3,5,6,7,7,7,8,2,2]
print("repeated numbers ",find_repeated_numbers(nums))
```
