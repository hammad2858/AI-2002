```python
#task1
import random

components = list("ABCDEFGHI")

system_state = {c: random.choice(["safe","vulnerable"]) for c in components}
print("initial system state : ")
print(system_state)

vulnerable_list = []

print("\n scanningSystem....")
for comp,state in system_state.items():
    if state =="vulnerable":
        print(f"{Warning } Component {comp} is vulnerable")
        vulnerable_list.append(comp)
    else:
        print(f"[ok] component{comp} is safe")
        
print("\n Patching Vulnerabilities...")
for comp in vulnerable_list:
    system_state[comp]="safe"
    print(f"Component {comp} patched")
    
print("\nFinal System State:")
print(system_state)
```

```python
#task2


servers = ["Server1", "Server2", "Server3", "Server4", "Server5"]
states = ["Underloaded", "Balanced", "Overloaded"]

server_load = {s: random.choice(states) for s in servers}

print("Initial Server Load:")
print(server_load)


overloaded = [s for s in servers if server_load[s] == "Overloaded"]
underloaded = [s for s in servers if server_load[s] == "Underloaded"]

print("\nBalancing Load...")
for o, u in zip(overloaded, underloaded):
    print(f"Moving tasks from {o} to {u}")
    server_load[o] = "Balanced"
    server_load[u] = "Balanced"

print("\nFinal Server Load:")
print(server_load)
```

```python
#task3
backup_tasks = [random.choice(["Completed","Failed"])for i in range (8)]
print("Initial Backup Status : ")
print(backup_tasks)

print("\n Retrying Failed Backup...")
for i in range(len(backup_tasks)):
    if backup_tasks[i]=="Failed":
        print(f"retrying Task{i}")
        backup_tasks[i] = "Completed"
        
print("\n final backup status :")
print(backup_tasks)
```

```python
#task4 utility based agent
components = list("ABCDEFGHI")
states = ["safe","low risk vulnerable","High risk Vulnerable"]

system  = {c: random.choice(states) for c in components}

print("Initial State:")
print(system)

print(" \n Scannning...")
for c,s in system.items():
    if s== "safe":
        print(f"{c}:safe")
    else:
        print(f"{c}: warning - {s}")
        
print("\n patching low risk vulnerabilities...")
for c in system:
    if system[c]=="Low Risk Vulnerable":
        system[c]="safe"
        print(f"{c} patched(low risk)")
    elif system[c] =="High Risk Vulnerable":
        print(f"{c} purchase premium service")
        
print("\n initial state :")
print(system)
```

```python
rooms = {
    'a': 'safe', 'b': 'safe', 'c': 'fire',
    'd': 'safe', 'e': 'fire', 'f': 'safe',
    'g': 'safe', 'h': 'safe', 'j': 'fire'
}

path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']

# Fixed grid layout
grid_layout = [
    ['a','b','c'],
    ['d','e','f'],
    ['g','h','j']
]

def display_env(env):
    print("\nEnvironment:")
    for row in grid_layout:
        print(" ".join(["fire" if env[r]=="fire" else "." for r in row]))

print("Initial Environment:")
display_env(rooms)

print("\nRobot Starting...")

for room in path:
    print(f"\nRobot entering room {room}")
    
    if rooms[room] == "fire":
        print("Fire detected! Extinguishing...")
        rooms[room] = "safe"
    else:
        print("Room safe")
    
    display_env(rooms)

print("\nFinal Environment:")
display_env(rooms)
```
