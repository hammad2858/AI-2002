```python
#prac 1 
class Light:
    def __init__(self, room):
        self.room = room
        self.status = "off"
        
    def turn_on(self):
        self.status = "ON"
        
    def turn_off(self):
        self.status = "OFF"
        
    def display(self):
        print(f"{self.room} light is {self.status}")

l1 = Light("Bedroom")
l2 = Light("Lounge")
l3  = Light("kitchen")

l1.turn_on()
l2.turn_off()
l3.turn_on()

l1.display()
l2.display()
l3.display()
```

```python
#prac task 2
class Staff:
    def __init__(self,name,id,dept):
        self.name = name
        self.id = id
        self.dept = dept
        
    def display(self):
        print(self.name,self.id,self.dept)
        
class Teacher(Staff):
    def __init__(self,name,id,dept,courses,salary):
        super().__init__(name,id,dept) # prev class constructor
        self.courses = courses
        self.salary = salary
        
    def teach(self):
        print(f"{self.name} is teaching {self.courses}")
              
    def display(self):
              print("name : ",self.name," ID : ",self.id," Department : ",self.dept," Courses : ", self.courses," salary : ",self.salary)
              
class Admin(Staff):
    def __init__(self,name,id,dept,role,working_hrs):
              super().__init__ (name,id,dept)
              self.role = role
              self.working_hrs = working_hrs
    def perform_task(self):
              print(f"{self.name} is performing admin task:")
              
    def display(self):
              print("name : ",self.name," ID : ",self.id," Department : ",self.dept," Role : ", self.role," Working Hours : ",self.working_hrs)

class ResearchAssistant(Staff):
    def __init__ (self,name,id,dept,research_topic,stipend):
              super().__init__(name,id,dept)
              self.research_topic = research_topic
              self.stipend = stipend
    
    def work_on_research(self):
              print(f"{self.name} is working on {self.research_topic}")
              
    def display(self):
              print("name : ",self.name," ID : ",self.id," Department : ",self.dept," research Topic : ", self.research_topic," Stipend : ",self.stipend)
              
              
t1 = Teacher("wasim rauf",892,"CS","Coal",200000)
a1 = Admin("Kashif",712,"Admin","Registrar",7 )
r1 = ResearchAssistant("sara",694,"Data Science","Big Data",50000)
              
t1.teach()
a1.perform_task()
r1.work_on_research()
              
              
t1.display()
a1.display()
r1.display()
              
              
              
```

```python
# prac task 3

class BankAccount:
    def __init__ (self,balance):
        self._balance = balance
        
    def deposit(self,amount):
        self._balance += amount
        
    def withdraw(self,amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient Balance")
            
    def get_Balance(self):
        return self._balance
acc1 = BankAccount(10000)
acc1.deposit(5000)
acc1.withdraw(2500)
acc1.get_Balance()
```

```python
#lab task 1

class SecurityAgent:
    def __init__ (self,id,name,status):
        self.id = id
        self.name = name
        self.status = status
        
    def display(self):
        print("\n ID : ,",self.id, " Name : ,",self.name," Status : ,", self.status)
        
class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        print("Monitoring Network traffic")
        
class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        print("Scanning files for malware")
        
class AutomationAgent(SecurityAgent):
    def run_automation(self):
        print("Running AI automation tasks")
        
f = FirewallAgent(1,"Firewall","Active")
md = MalwareDetectionAgent(2,"Malware Detector","Active")
Auto = AutomationAgent(3,"Automation","Active")

f.monitor_traffic()
md.scan_files()
Auto.run_automation()

f.display()
md.display()
Auto.display()
```

```python
#task 2
class Threat:
    def __init__ (self,id,name,severity):
        self.id=id
        self.name=name
        self.severity=severity
    def display(self):
        print("\n threat ID : ,",self.id, "  threat Name : ,",self.name," Severity : ,", self.severity)
class Phishing(Threat):
    def analyze_email(self):
        print("Analyzing Phishing emails")

class Ransomware(Threat):
    def scan_files(self):
        print("scanning file system")
        
class Botnet(Threat):
    def detect_traffic(self):
        print("Detecting botnet traffic")
        
p = Phishing(1,"Phishing","High")
r = Ransomware(2,"Ransomware","Low")
b = Botnet(3,"Botnet","Critical")

p.analyze_email()
r.scan_files()
b.detect_traffic()


p.display()
r.display()
b.display()
```

```python
#task 3 

class ResponseAgent:
    def execute_response(self):
        pass
    
class AlertAgent(ResponseAgent):
    def execute_response(self):
        print("Sending alert notification")
        
class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("Blocking Malicious Activity")
        
class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("Recovering System")
        
agents = [AlertAgent(),BlockAgent(),RecoverAgent()]

for agent in agents:
    agent.execute_response()
```

```python
#task4
class Employee:
    def work(self):
        pass
    
class Manager(Employee):
    def work(self):
        print(" Managing Team ")
        
class Developer(Employee):
    def work(self):
        print(" Coding ")
        
class Designer(Employee):
    def work(self):
        print(" Designing UI ")
        
employees = [Manager(),Developer(),Designer()]

for empl in employees : 
    empl.work()
```
